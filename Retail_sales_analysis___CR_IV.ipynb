```python
# Cristian Rivas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#IMPORTANTE: al ser esta tarea la continuación de otra, dejaré tal cual está la primera parte.

# 1. Cargar los Datos
# Ruta del archivo CSV
archivo_csv = 'retail_sales_dataset.csv'

# Cargar los datos en un DataFrame
df = pd.read_csv(archivo_csv)

# Mostrar las primeras 10 filas para confirmar que los datos se han cargado correctamente
print("Primeras 10 filas del DataFrame:")
print(df.head(10))
print("\n" + "="*50 + "\n")
```

```python
# 2. Exploración Inicial de los Datos
# Mostrar las últimas 5 filas del DataFrame
print("Últimas 5 filas del DataFrame:")
print(df.tail(5))
print("\n" + "="*50 + "\n")

# Utilizar el método info() para obtener información general sobre el DataFrame
print("Información general del DataFrame:")
df.info()
print("\n" + "="*50 + "\n")

# Generar estadísticas descriptivas del DataFrame utilizando el método describe()
print("Estadísticas descriptivas del DataFrame:")
print(df.describe())
print("\n" + "="*50 + "\n")
```

```python
# 3. Inspección de los Datos
# Inspeccionar los tipos de datos de cada columna utilizando el atributo dtypes
print("Tipos de datos de cada columna:")
print(df.dtypes)
print("\n" + "="*50 + "\n")

# Contar los valores únicos en la columna Producto utilizando el método value_counts()
print("Conteo de valores únicos en la columna 'Product Category':")
print(df['Product Category'].value_counts())
print("\n" + "="*50 + "\n")

#Segun entiendo, las instrucciones no están del todo acertadas, voy a hacer lo que me parezca
# Mostrar todos los valores únicos en la columna Customer ID utilizando el método unique()
print("Valores únicos en la columna 'Customer ID':")
print(df['Customer ID'].unique())
print("\n" + "="*50 + "\n")
```

```python
# 4. Filtrado de Datos
# Filtrar el DataFrame para mostrar solo las filas donde las ventas (Total Amount) sean mayores a 50
print("Filas donde 'Total Amount' es mayor a 50:")
df_sales_above_50 = df[df['Total Amount'] > 50]
print(df_sales_above_50)
print("\n" + "="*50 + "\n")

# Filtrar el DataFrame para mostrar solo las filas donde el precio (Price per Unit) sea menor a 0.5
print("Filas donde 'Price per Unit' es menor a 0.5:")
df_price_below_05 = df[df['Price per Unit'] < 0.5]
print(df_price_below_05)
print("\n" + "="*50 + "\n")

# Utilizando el método query(), filtrar el DataFrame para mostrar las filas donde el producto sea 'Beauty'
# y las ventas sean mayores a 30
print("Filas donde 'Product Category' es 'Beauty' y 'Total Amount' es mayor a 30:")
df_beauty_above_30 = df.query("`Product Category` == 'Beauty' and `Total Amount` > 30")
print(df_beauty_above_30)
print("\n" + "="*50 + "\n")

```

```python
# 5. Slicing de Datos
# Seleccionar y mostrar solo las columnas Producto y Ventas del DataFrame
print("Columnas 'Product Category' y 'Total Amount':")
print(df[['Product Category', 'Total Amount']])
print("\n" + "="*50 + "\n")

# Utilizando loc[], seleccionar y mostrar las filas de la 5 a la 10 (inclusive) y las columnas Producto y Tienda
print("Filas de la 5 a la 10 y columnas 'Product Category' y 'Customer ID':")
print(df.loc[5:10, ['Product Category', 'Customer ID']])
print("\n" + "="*50 + "\n")

# Utilizando iloc[], seleccionar y mostrar las primeras 5 filas y las primeras 3 columnas del DataFrame
print("Primeras 5 filas y primeras 3 columnas del DataFrame:")
print(df.iloc[:5, :3])
```

```python
# Paso 1: Transformación de datos

# Crear una nueva columna 'Normalized Sales' normalizando las ventas (Total Amount) entre 0 y 1
df['Normalized Sales'] = (df['Total Amount'] - df['Total Amount'].min()) / (df['Total Amount'].max() - df['Total Amount'].min())
print("Columna 'Normalized Sales' creada y normalizada entre 0 y 1.")

# Crear una nueva columna 'Sales Category' que clasifique las ventas en categorías ('Baja', 'Media', 'Alta')
def clasificar_ventas(monto):
    if monto < 100:
        return 'Baja'
    elif 100 <= monto < 500:
        return 'Media'
    else:
        return 'Alta'

df['Sales Category'] = df['Total Amount'].apply(clasificar_ventas)
print("\nColumna 'Sales Category' creada para clasificar las ventas ('Baja', 'Media', 'Alta').")

# Mostrar las primeras filas del DataFrame después de la transformación
print("\nPrimeras filas del DataFrame transformado:")
print(df.head())
```

```python
# Paso 2: Agrupación y agregación

# Agrupar por 'Product Category' y calcular el total de ventas, promedio y número de transacciones
print("\nAgrupación por 'Product Category': Total de ventas, promedio y número de transacciones.")
agg_product_category = df.groupby('Product Category').agg(
    Total_Sales=('Total Amount', 'sum'),
    Average_Sales=('Total Amount', 'mean'),
    Transaction_Count=('Transaction ID', 'count')
)
print(agg_product_category)

# Agrupar por 'Sales Category' y 'Product Category' y calcular el promedio de ventas normalizadas
print("\nAgrupación por 'Sales Category' y 'Product Category': Promedio de ventas normalizadas.")
agg_sales_category = df.groupby(['Sales Category', 'Product Category']).agg(
    Average_Normalized_Sales=('Normalized Sales', 'mean')
)
print(agg_sales_category)

# Mostrar las primeras filas del DataFrame transformado para verificar la agregación
print("\nPrimeras filas del DataFrame después de la agregación:")
print(df.head())
```

```python
# Paso 3: Análisis personalizado con apply

# Crear una columna 'Group Mean Deviation' para calcular la desviación de cada venta respecto al promedio de su categoría de producto
def calcular_desviacion(row, group_means):
    return row['Total Amount'] - group_means[row['Product Category']]

# Calcular las medias de ventas por categoría de producto
group_means = df.groupby('Product Category')['Total Amount'].mean().to_dict()

# Aplicar la función personalizada al DataFrame
df['Group Mean Deviation'] = df.apply(lambda row: calcular_desviacion(row, group_means), axis=1)
print("\nColumna 'Group Mean Deviation' creada con la desviación respecto a la media de su grupo.")

# Mostrar las primeras filas del DataFrame para verificar
print("\nPrimeras filas del DataFrame después del análisis personalizado:")
print(df.head())
```

```python
# Paso 4: Documentación

# Resumen de las operaciones realizadas y su propósito:

# 1. Transformación de Datos:
#    - Se creó la columna 'Normalized Sales' para normalizar las ventas (Total Amount) en un rango de 0 a 1.
#      Esto permite realizar comparaciones independientes de la escala de valores.
#    - Se creó la columna 'Sales Category' para clasificar las ventas como 'Baja', 'Media' o 'Alta'.
#      Esto facilita el análisis categórico y segmentación de las ventas.

# 2. Agrupación y Agregación:
#    - Se agruparon los datos por 'Product Category' para calcular:
#         - Total de ventas (Total Sales).
#         - Promedio de ventas (Average Sales).
#         - Número de transacciones (Transaction Count).
#      Esto proporciona información resumida clave por categoría de producto.
#    - Se agruparon los datos por 'Sales Category' y 'Product Category' para calcular
#      el promedio de ventas normalizadas (Average Normalized Sales), permitiendo observar tendencias por grupo.

# 3. Análisis Personalizado con apply:
#    - Se creó la columna 'Group Mean Deviation' que calcula la desviación de cada venta respecto a
#      la media de su categoría de producto. Esto permite identificar transacciones que están por
#      encima o por debajo del promedio de su grupo.
```

```python

# Configuración del tamaño de los gráficos
plt.figure(figsize=(16, 12))

# Histograma para 'Total Amount'
plt.subplot(2, 3, 1)
sns.histplot(df['Total Amount'], kde=True, color='skyblue')
plt.title('Histograma de Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frecuencia')

# Boxplot para 'Total Amount'
plt.subplot(2, 3, 4)
sns.boxplot(x=df['Total Amount'], color='skyblue')
plt.title('Boxplot de Total Amount')
plt.xlabel('Total Amount')

# Histograma para 'Price per Unit'
plt.subplot(2, 3, 2)
sns.histplot(df['Price per Unit'], kde=True, color='orange')
plt.title('Histograma de Price per Unit')
plt.xlabel('Price per Unit')
plt.ylabel('Frecuencia')

# Boxplot para 'Price per Unit'
plt.subplot(2, 3, 5)
sns.boxplot(x=df['Price per Unit'], color='orange')
plt.title('Boxplot de Price per Unit')
plt.xlabel('Price per Unit')

# Histograma para 'Normalized Sales'
plt.subplot(2, 3, 3)
sns.histplot(df['Normalized Sales'], kde=True, color='green')
plt.title('Histograma de Normalized Sales')
plt.xlabel('Normalized Sales')
plt.ylabel('Frecuencia')

# Boxplot para 'Normalized Sales'
plt.subplot(2, 3, 6)
sns.boxplot(x=df['Normalized Sales'], color='green')
plt.title('Boxplot de Normalized Sales')
plt.xlabel('Normalized Sales')

# Mostrar gráficos
plt.tight_layout()
plt.show()
```

```python

# Convertir la columna 'Date' a tipo datetime para facilitar la manipulación
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar por fecha y calcular la suma de las ventas totales para cada día
sales_trend = df.groupby('Date')['Total Amount'].sum()

# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))
plt.plot(sales_trend.index, sales_trend.values, marker='o', linestyle='-', color='b')
plt.title('Tendencia de Ventas Totales a lo Largo del Tiempo', fontsize=14)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Ventas Totales', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python


# Gráfico de dispersión: Total Amount vs Price per Unit
plt.figure(figsize=(14, 6))

# Subplot 1: Total Amount vs Price per Unit
plt.subplot(1, 2, 1)
plt.scatter(df['Price per Unit'], df['Total Amount'], alpha=0.7, c='orange', edgecolors='k')
plt.title('Relación: Total Amount vs Price per Unit')
plt.xlabel('Price per Unit')
plt.ylabel('Total Amount')
plt.grid(True)

# Subplot 2: Normalized Sales vs Age
plt.subplot(1, 2, 2)
plt.scatter(df['Age'], df['Normalized Sales'], alpha=0.7, c='green', edgecolors='k')
plt.title('Relación: Normalized Sales vs Age')
plt.xlabel('Age')
plt.ylabel('Normalized Sales')
plt.grid(True)

plt.tight_layout()
plt.show()
```
